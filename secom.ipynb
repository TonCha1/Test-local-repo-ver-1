```python
import pandas as pd
url_of_label = 'https://raw.githubusercontent.com/TonCha1/Test-local-repo-ver-1/main/secom_labels.data'
url_of_data = 'https://raw.githubusercontent.com/TonCha1/Test-local-repo-ver-1/main/secom.data'
label = pd.read_csv(url_of_label, sep=' ')
data = pd.read_csv(url_of_data, sep=' ')
print(label.head(3),data.head(3))
print(label.shape,data.shape)
```

```python
secom = pd.concat([label, data], ignore_index=True, axis=1)
secom.shape
secom.head(3)
```
